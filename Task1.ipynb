```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_excel("/content/P_Data_Extract_From_World_Development_Indicators.xlsx")
df1
```

```python
c = df1.groupby("Country Code")
c
```

```python
c = c.get_group("IND")
c
```

```python
c = c.drop('Series Code', axis=1)
c
```

**CO2 emissions metric tonnes per capita**


```python
c.replace('..', np.nan, inplace=True)
years = c.columns[4:]
values = c.iloc[1, 4:].astype(float)
plt.figure(figsize=(10, 6))
plt.plot(years, values, marker='o', linestyle='-', color='m')
plt.title('CO2 EMISSIONS METRIC TONNES PER CAPITA')
plt.xlabel('Year')
plt.ylabel('CO2 emissions metrictonnes per capita')
plt.grid(True)
plt.xticks(rotation = 45 ,ha ="right")
plt.show()
```

```python
c.replace('..', np.nan, inplace=True)

years = c.columns[4:]
values = c.iloc[14, 4:].astype(float)

plt.figure(figsize=(10, 6))
plt.bar(years,values,color='darkcyan')
plt.title('GDP(Current) OF INDIA IN VARIOUS YEARS')
plt.xlabel('Year')
plt.ylabel('(GDP IN US $) * 10^12')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
c.replace('..', np.nan, inplace=True)
years = c.columns[4:]
values = c.iloc[15, 4:].astype(float)
plt.figure(figsize=(10, 6))
plt.title('GDP GROWTH (%) of INDIA OVER THE YEARS')
plt.plot(years, values, marker='o', linestyle='-', color='g')
plt.xlabel('Year')
plt.ylabel('GDP ANNUAL GROWTH (%)')
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
c.replace('..', np.nan, inplace=True)
years = c.columns[7:]
values = c.iloc[10, 7:].astype(float)
plt.figure(figsize=(10, 6))
plt.title('EXPORTS OF GOODS AND SERVICES (% OF GDP)')
custom_colors = ['coral', 'darkcyan', 'dimgrey', 'steelblue', 'indianred']
plt.pie(values,labels = years,autopct ='%1.1f%%',colors=custom_colors,startangle = 30)
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python

fig, ax1 = plt.subplots(figsize=(12, 6))  # Adjust the figure size as needed
color = 'tab:red'
ax1.set_xlabel('Year')
ax1.set_ylabel('EXPORTS OF GOODS AND SERVICES (% OF GDP)', color=color)
ax1.plot(c.columns[3:], c.iloc[9, 3:], color=color, marker='*', label='Exports Of Goods And Services (% of GDP)')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('IMPORTS OF GOODS AND SERVICES (% OF GDP)', color=color)
ax2.plot(c.columns[3:], c.iloc[23, 3:], color=color, marker='o', label='Imports Of Goods And Services (% Of GDP)')
ax2.tick_params(axis='y', labelcolor=color)

# Increase space between subplots to avoid overlapping labels
fig.tight_layout()

# Display grid
plt.grid(True)

# Display legend
fig.legend(loc="lower right", bbox_to_anchor=(0.6, 0.8))


# Show the plot
plt.show()
```
